```python
import numpy as np
import matplotlib.pyplot as plt
from math import fabs
```

```python
np.random.seed(np.random.randint(0, 1_000_000))
num_points = np.random.randint(0, 1_000)
a = np.random.uniform(0, 10)
b = np.random.uniform(0, 10)
noise = np.random.normal(0, 1, num_points)
X = np.random.uniform(0, 5, num_points)
Y = a * X + b + noise

print("Size: \t", num_points)
```

```python
plt.scatter(X, Y, marker = '.', color = 'r')
plt.plot(X, a * X + b)
plt.xlabel('X')
plt.ylabel('Y')
plt.grid()
plt.show()
```

```python
# init parameters
pred_a = 0
pred_b = 0
# learning speed
L = 0.08

x_size = float(len(X))

D_a = 0
D_b = 0
# linear regression
while True:
    D_a_old = D_a
    D_b_old = D_b
    Y_pred = pred_a * X + pred_b
    D_a = (-2 / x_size) * sum(X * (Y - Y_pred))
    D_b = (-2 / x_size) * sum(Y - Y_pred)
    pred_a = pred_a - L * D_a
    pred_b = pred_b - L * D_b
    if fabs(D_a - D_a_old) < 0.0001 or (fabs(D_b - D_b_old) < 0.0001):
        break;
        
Y_pred = pred_a * X + pred_b
```

```python
plt.scatter(X, Y, marker = '.', color = 'r')
plt.plot(X, a * X + b, color='blue')
plt.plot(X, Y_pred, color='green')

plt.xlabel('X')
plt.ylabel('Y')
plt.grid()
plt.show()

print ('Original a: \t', a)
print ('Predicted a: \t', pred_a)
print ('Original b: \t', b)
print ('Predicted b: \t', pred_b)
```
